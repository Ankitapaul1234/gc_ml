# **Ordinal Encoding**

```python
df_O=pd.DataFrame({"size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df_O
```

# **using sckitlearn**

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df_O[["size"]])
```

```python
df_O["OE"] = oe.transform(df_O[["size"]])
```

```python
df_O
```

# **using map function**

```python
ord_data1={"s":0,"m":1,"l":2,"xl":3}
```

```python
ord_data1={"s":10,"m":11,"l":12,"xl":13}
```

```python
df_O["map"] = df_O["size"].map(ord_data1)
```
