```python
points_df = pd.DataFrame(
    {
        '1st point' : [(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point': [(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)
points_df
```

```python
def euclidean(row):
  point1 = row['1st point']
  point2 = row['2nd point']
  return np.sqrt((point1[0]-point2[0])**2 + (point1[1]-point2[1])**2)
```

```python
points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
```

```python
#isin  --> It is commonly used for filtering rows based on multiple values in a column.
# Sample DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
})

df
```

```python
# Filtering rows where 'City' is either 'New York' or 'Chicago'
filtered_df = df[df['City'].isin(['New York', 'Chicago'])]
print(filtered_df)
```

```python
#using multiple col
# Filtering where 'City' is in ['New York', 'Chicago'] and 'Age' is in [30, 40]
filtered_df = df[df.isin({'City': ['New York', 'Chicago'], 'Age': [30, 40]})]
print(filtered_df)
```

```python
#nlargest() and nsmallest() --> The .nlargest(n, columns) and .nsmallest(n, columns) methods return the top or bottom n rows from a DataFrame or Series based on a specific column.
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Score': [85, 92, 78, 90, 88]
})
df
```

```python
# Get the top 3 highest scores
top_3 = df.nlargest(3, 'Score')
print(top_3)
```

```python
# Get the bottom 2 lowest scores
bottom_2 = df.nsmallest(2, 'Score')
print(bottom_2)
```

```python
#insert --> The .insert() method allows you to insert a new column at a specific position in a DataFrame
# Insert a new column "Rank" at index 1
df.insert(1, 'Rank', [3, 1, 5, 2, 4])
print(df)
```

```python
#copy --> method creates a copy of a DataFrame or Series to avoid modifying the original data.
#Parameters
#deep=True (default): Creates a full copy (data and structure).
#deep=False: Creates a shallow copy (changes to copied DataFrame affect the original).
####Example: Deep Copy (Safe)

df_copy = df.copy()  # Default is deep copy
df_copy['Score'][0] = 99  # Change value in copy

print(df)        # Original remains unchanged
print(df_copy)   # Copy is modified
```

```python
#Example: Shallow Copy (Not Safe)

df_shallow = df.copy(deep=False)
df_shallow['Score'][0] = 99  # Changes will reflect in df!

print(df)  # Original is affected!
```
